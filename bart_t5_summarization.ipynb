```python
#loading both models
from transformers import BartTokenizer, BartForConditionalGeneration
from transformers import T5Tokenizer, T5ForConditionalGeneration
print('Loading BART...')
bart_tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
bart_model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
print('BART Loaded!')

print('Loading T5...')
t5_tokenizer = T5Tokenizer.from_pretrained('google/flan-t5-base')
t5_model = T5ForConditionalGeneration.from_pretrained('google/flan-t5-base')
print('T5 Loaded!')

print('Both models successfully loaded!')
```

```python
#Test text
test_text = '''
Messenger RNA (mRNA) vaccines work by introducing a piece of genetic code
into the body that instructs cells to produce a protein similar to one found
on the surface of a virus. The immune system recognizes this protein as foreign
and builds a defense against it. If the person is later exposed to the actual
virus, the immune system is already prepared to fight it. Unlike traditional
vaccines, mRNA vaccines do not contain any live or inactivated virus particles.
The mRNA itself breaks down quickly in the body after the immune response is triggered.
'''

# BART summary
print('=== BART SUMMARY ===')
bart_inputs = bart_tokenizer(test_text, return_tensors='pt', max_length=1024, truncation=True)
bart_output = bart_model.generate(**bart_inputs, max_length=150, min_length=30)
bart_summary = bart_tokenizer.decode(bart_output[0], skip_special_tokens=True)
print(bart_summary)

# T5 summary
print('\n=== T5 SUMMARY ===')
t5_inputs = t5_tokenizer(test_text, return_tensors='pt', max_length=512, truncation=True)
t5_output = t5_model.generate(**t5_inputs, max_length=150, min_length=30)
t5_summary = t5_tokenizer.decode(t5_output[0], skip_special_tokens=True)
print(t5_summary)
```
